```python
import pandas as pd
df = pd.read_csv('country_vaccinations.csv')
df_population = pd.read_csv('world_population.csv')
#cambiamos el nombre de la columna
df_population = df_population.rename(columns={'Country/Territory':'country'})
```

```python
#1. Promedio de vacunas administradas
vaccinations_mean = df.groupby('country')['daily_vaccinations'].mean()
vaccinations_mean
```

```python
#Añadir columna población por cada país
df_merge = pd.merge(df, df_population, on="country", how="left")

#rellenamos con 0 los dias que tienen NaN en daily vaccination
df_merge['daily_vaccinations'] = df['daily_vaccinations'].fillna(0)
df_merge
```

```python
#2. tasa de vacunación por cada 100,000 habitantes
df_merge['vaccination_rate_100000'] = df_merge.apply(lambda row: ((row['daily_vaccinations'] / row['2022 Population'])  * 100000), axis=1)
df_merge

```

```python
#3.Paises con más del 70% de la población vacunada
#df_merge[df_merge['vaccination_rate_100000'] > 0]
df_merge.groupby('country')['daily_vaccinations'].sum()/(df_merge.groupby('country')['2022 Population'].sum())
```
